```python
from tkinter import *
root = Tk()
root.title("Simple Calculator")

e = Entry(root, width=55, borderwidth=4, bg="OldLace")
e.grid(row=0, column=0, columnspan=5, padx=20, pady=20)


def button_click(*args):
    current = e.get()
    e.delete(0,END)
    e.insert(0,str(current) + str(*args))
    
def button_clear():
    e.delete(0,END)
    
def button_back():
    e.delete(len(e.get())-1)
    
def button_add():
    first_num = e.get()
    global f_num
    global math
    math = "add"
    f_num = int(first_num)
    e.delete(0,END)
    
def button_sub():
    first_num = e.get()
    global f_num
    global math
    math = "sub"
    f_num = int(first_num)
    e.delete(0,END)

def button_mul():
    first_num = e.get()
    global f_num
    global math
    math = "mul"
    f_num = int(first_num)
    e.delete(0,END)

def button_div():
    first_num = e.get()
    global f_num
    global math
    math = "div"
    f_num = int(first_num)
    e.delete(0,END)
    
def button_eq():
    second_num = e.get()
    e.delete(0,END)
    
    if math == "add":
        e.insert(0,f_num + int(second_num))

    if math == "sub":
        e.insert(0,f_num - int(second_num))
        
    if math == "mul":
        e.insert(0,f_num * int(second_num))
        
    if math == "div":
        e.insert(0,f_num / int(second_num))
        
        
button_1 = Button(root, text="1", padx=40, pady=20, command=lambda: button_click(1), bg ="GhostWhite")
button_2 = Button(root, text="2", padx=40, pady=20, command=lambda: button_click(2), bg ="GhostWhite")
button_3 = Button(root, text="3", padx=40, pady=20, command=lambda: button_click(3), bg ="GhostWhite")
button_4 = Button(root, text="4", padx=40, pady=20, command=lambda: button_click(4), bg ="GhostWhite")
button_5 = Button(root, text="5", padx=40, pady=20, command=lambda: button_click(5), bg ="GhostWhite")
button_6 = Button(root, text="6", padx=40, pady=20, command=lambda: button_click(6), bg ="GhostWhite")
button_7 = Button(root, text="7", padx=40, pady=20, command=lambda: button_click(7), bg ="GhostWhite")
button_8 = Button(root, text="8", padx=40, pady=20, command=lambda: button_click(8), bg ="GhostWhite")
button_9 = Button(root, text="9", padx=40, pady=20, command=lambda: button_click(9), bg ="GhostWhite")
button_0 = Button(root, text="0", padx=40, pady=20, command=lambda: button_click(0), bg="GhostWhite")
button_back = Button(root, text="โซ", padx=40, pady=20, command=button_back,bg="PaleVioletRed2")
button_dot = Button(root, text=".", padx=40, pady=20,bg="GhostWhite")
button_ADD = Button(root, text="+", padx=40, pady=20, command=button_add, bg ="LightSlateGray")
button_sub = Button(root, text="-", padx=40, pady=20, command=button_sub, bg ="LightSlateGray")
button_mul = Button(root, text="x", padx=40, pady=20, command=button_mul, bg ="LightSlateGray")
button_div = Button(root, text="รท", padx=40, pady=20, command=button_div, bg ="LightSlateGray")
button_eq = Button(root, text="=", padx=87, pady=20, command=button_eq, bg="LightSalmon")
button_ce = Button(root, text="CE", padx=87, pady=20, command=button_clear, bg="LightSalmon")


button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)
button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)
button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)
button_back.grid(row=1, column=3)
button_0.grid(row=2, column=3)
button_dot.grid(row=3, column=3)
button_mul.grid(row=4, column=0)
button_div.grid(row=4, column=1)
button_ADD.grid(row=4, column=2)
button_sub.grid(row=4, column=3)
button_eq.grid(row=5, column=0, columnspan=2)
button_ce.grid(row=5, column=2, columnspan=2)

root.mainloop()
```
