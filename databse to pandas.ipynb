```python
import mysql.connector
import pandas as pd
```

```python
mydb = mysql.connector.connect(host="localhost",
user="root",
password="2002",
database="seller_db")
```

```python
mycursor = mydb.cursor()
mycursor.execute('select * from book_details')
data = mycursor.fetchall()
for i in data:
    print(i)
```

```python
id=[]
year=[]
semester=[]
branch=[]
subject=[]
phone=[]
email=[]
for i in data:
    id.append(i[0])
    year.append(i[1])
    semester.append(i[2])
    branch.append(i[3])
    subject.append(i[4])
    phone.append(i[5])
    email.append(i[6])
    
d = {'id':id, 'year':year, 'semester':semester, 'branch':branch, 'subject':subject, 'phone':phone, 'email':email}
df = pd.DataFrame(d)
df
```
